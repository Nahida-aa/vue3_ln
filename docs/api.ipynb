### 用户个人信息

#### 请求方式
```sh
GET
```
#### 请求参数
```sh
无
```
#### 响应示例
```json
{
	"errCode": 0,
	"errMsg": "查询成功",
	"data": {
		"IP": "100.100.100.101",
		"address": {
			"country": "中国",
			"province": "山东",
			"city": "济南"
		},
		"downloadSize": 5,
		"scoreSize": 8
	},
	"timeCost": 132
}
```
| name | type | description |

| name | type | description |

该接口并不是实时的数据统计，比如下载次数和评分个数，这个是有缓存的5分钟，这是为了减少运算，提高查询效率。
